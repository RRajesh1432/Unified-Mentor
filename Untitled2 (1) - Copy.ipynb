```python
import os
import json
import numpy as np
import librosa
import torch
import torch.nn as nn

from torch.utils.data import Dataset, DataLoader
from transformers import AutoProcessor, AutoModelForAudioClassification
from sklearn.model_selection import train_test_split
```

```python
import os


URBAN_ROOT = "/content/drive/MyDrive/urban8k_reduced"

print(os.listdir(URBAN_ROOT))


# fsd50k path obtained above
FSD50K_ROOT = path

MODEL_NAME = "MIT/ast-finetuned-audioset-10-10-0.4593"

SR = 16000
CROP_SEC = 2.0

BATCH_SIZE = 4
EPOCHS = 12
LR = 1e-4

LABELS = ["gunshot", "siren", "glass_break", "distress_call", "normal"]
LABEL2ID = {k:i for i,k in enumerate(LABELS)}
ID2LABEL = {v:k for k,v in LABEL2ID.items()}

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(DEVICE)
```

```python
LABEL_MAP = {

    "gunshot": [
        "Gunshot, gun fire",
        "Gunshot, single shot",
        "Gunshot, machine gun",
        "Rifle fire",
        "Pistol shot",
        "Shotgun blast",
        "Automatic firearm"
    ],

    "siren": [
        "Siren",
        "Civil defense siren",
        "Ambulance siren",
        "Police siren",
        "Fire engine siren",
        "Emergency vehicle"
    ],

    "glass_break": [
        "Glass break",
        "Breaking glass",
        "Shattering glass",
        "Window breaking",
        "Bottle breaking"
    ],

    "distress_call": [
        "Scream",
        "Female scream",
        "Male scream",
        "Child scream",
        "Shout",
        "Yell",
        "Help",
        "Crying, sobbing",
        "Panic scream",
        "Distress vocalization"
    ]
}
```

```python
PRIORITY = ["gunshot","glass_break","distress_call","siren"]
```

```python
# find csv metadata
csv_files = []
for root,_,files in os.walk(FSD50K_ROOT):
    for f in files:
        if f.endswith(".csv"):
            csv_files.append(os.path.join(root,f))

csv_files
```

```python
import pandas as pd

# choose dev set
meta_csv = [c for c in csv_files if "dev" in c.lower()][0]
print(meta_csv)

fsd_meta = pd.read_csv(meta_csv)
fsd_meta.head()
```

```python
audio_index = {}

for root,_,files in os.walk(FSD50K_ROOT):
    for f in files:
        if f.lower().endswith(".wav"):
            audio_index[f] = os.path.join(root,f)

len(audio_index)
```

```python
def map_fsd_labels(label_string):

    labs = [l.strip() for l in label_string.split(",")]

    for tgt in PRIORITY:
        for src in LABEL_MAP[tgt]:
            if src in labs:
                return tgt

    return "normal"
```

```python
fsd_files = []
fsd_labels = []

missing = 0

for _,row in fsd_meta.iterrows():

    fname = row["fname"]
    labels = row["labels"]

    if fname not in audio_index:
        missing += 1
        continue

    tgt = map_fsd_labels(labels)

    fsd_files.append(audio_index[fname])
    fsd_labels.append(LABEL2ID[tgt])

print("FSD used:",len(fsd_files)," missing:",missing)
```

```python
urban_files = []
urban_labels = []

for cls in LABELS:
    folder = os.path.join(URBAN_ROOT,cls)
    for f in os.listdir(folder):
        if f.lower().endswith((".wav",".mp3",".flac")):
            urban_files.append(os.path.join(folder,f))
            urban_labels.append(LABEL2ID[cls])

print("Urban:",len(urban_files))
```

```python
all_files  = urban_files + fsd_files
all_labels = urban_labels + fsd_labels

train_f, val_f, train_y, val_y = train_test_split(
    all_files,
    all_labels,
    test_size=0.15,
    stratify=all_labels,
    random_state=42
)

len(train_f), len(val_f)
```

```python
def peak_crop(y, sr, crop_sec=2.0):

    win = int(sr * 0.05)
    hop = int(sr * 0.01)

    rms = librosa.feature.rms(y=y, frame_length=win, hop_length=hop)[0]
    peak = np.argmax(rms)

    center = int(peak * hop)
    half = int((crop_sec * sr) // 2)

    start = max(0, center-half)
    end = start + int(crop_sec*sr)

    if end > len(y):
        end = len(y)
        start = max(0, end-int(crop_sec*sr))

    seg = y[start:end]

    if len(seg) < int(crop_sec*sr):
        seg = np.pad(seg,(0,int(crop_sec*sr)-len(seg)))

    return seg
```

```python
class AudioDataset(Dataset):

    def __init__(self, files, labels, train=True):
        self.files = files
        self.labels = labels
        self.train = train

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):

        y,_ = librosa.load(self.files[idx], sr=SR, mono=True)
        label = self.labels[idx]

        seg_len = int(SR*CROP_SEC)

        if self.train:
            if len(y) > seg_len:
                s = np.random.randint(0,len(y)-seg_len+1)
                y = y[s:s+seg_len]
            else:
                y = np.pad(y,(0,seg_len-len(y)))

            y = y + 0.002*np.random.randn(len(y))

        else:
            y = peak_crop(y,SR,CROP_SEC)

        if len(y) < seg_len:
            y = np.pad(y,(0,seg_len-len(y)))

        return y.astype(np.float32), label
```

```python
processor = AutoProcessor.from_pretrained(MODEL_NAME)
```

```python
def collate_fn(batch):

    audios = [b[0] for b in batch]
    labels = torch.tensor([b[1] for b in batch])

    inp = processor(
        audios,
        sampling_rate=SR,
        return_tensors="pt",
        padding=True
    )

    return inp["input_values"], labels
```

```python
train_ds = AudioDataset(train_f,train_y,train=True)
val_ds   = AudioDataset(val_f,val_y,train=False)

train_loader = DataLoader(
    train_ds,
    batch_size=BATCH_SIZE,
    shuffle=True,
    collate_fn=collate_fn,
    num_workers=2,
    pin_memory=True
)

val_loader = DataLoader(
    val_ds,
    batch_size=BATCH_SIZE,
    shuffle=False,
    collate_fn=collate_fn,
    num_workers=2,
    pin_memory=True
)
```

```python
class ASTThreatModel(nn.Module):

    def __init__(self,nc):
        super().__init__()

        self.ast = AutoModelForAudioClassification.from_pretrained(
            MODEL_NAME
        )

        h = self.ast.config.hidden_size
        self.classifier = nn.Linear(h,nc)

    def forward(self,x):

        out = self.ast(
            input_values=x,
            output_hidden_states=True
        )

        z = out.hidden_states[-1].mean(dim=1)
        return self.classifier(z)
```

```python
model = ASTThreatModel(len(LABELS)).to(DEVICE)

for p in model.ast.parameters():
    p.requires_grad = False

try:
    for p in model.ast.audio_spectrogram_transformer.encoder.layer[-1].parameters():
        p.requires_grad = True
except:
    for p in model.ast.base_model.encoder.layer[-1].parameters():
        p.requires_grad = True
```

```python
optimizer = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model.parameters()),
    lr=LR
)

criterion = nn.CrossEntropyLoss()
scaler = torch.cuda.amp.GradScaler(enabled=(DEVICE=="cuda"))
```

```python
def train_one_epoch():

    model.train()
    tot,cor,ls = 0,0,0

    for x,y in train_loader:

        x = x.to(DEVICE)
        y = y.to(DEVICE)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast(enabled=(DEVICE=="cuda")):
            out = model(x)
            loss = criterion(out,y)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        p = out.argmax(1)
        tot += y.size(0)
        cor += (p==y).sum().item()
        ls  += loss.item()

    return ls/len(train_loader), cor/tot
```

```python
@torch.no_grad()
def validate():

    model.eval()
    tot,cor,ls = 0,0,0

    for x,y in val_loader:

        x = x.to(DEVICE)
        y = y.to(DEVICE)

        with torch.cuda.amp.autocast(enabled=(DEVICE=="cuda")):
            out = model(x)
            loss = criterion(out,y)

        p = out.argmax(1)
        tot += y.size(0)
        cor += (p==y).sum().item()
        ls  += loss.item()

    return ls/len(val_loader), cor/tot
```

```python
SAVE_DIR = "/content/drive/MyDrive/best_fsd50k_threat_model"
os.makedirs(SAVE_DIR,exist_ok=True)

best = 0

for e in range(EPOCHS):

    tl,ta = train_one_epoch()
    vl,va = validate()

    print(f"Epoch {e+1:02d} | train acc {ta:.4f} | val acc {va:.4f}")

    if va > best:
        best = va
        torch.save(
            model.state_dict(),
            os.path.join(SAVE_DIR,"best_model.pt")
        )

with open(os.path.join(SAVE_DIR,"label_map.json"),"w") as f:
    json.dump(LABEL2ID,f,indent=2)

print("Best validation accuracy:",best)
```

```python
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt

from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
from sklearn.preprocessing import label_binarize

# -----------------------------------
# Collect predictions
# -----------------------------------

y_true = []
y_pred = []
y_prob = []

model.eval()

with torch.no_grad():
    for x, y in val_loader:

        x = x.to(DEVICE)

        logits = model(x)
        probs = torch.softmax(logits, dim=1)

        y_true.extend(y.numpy())
        y_pred.extend(probs.argmax(dim=1).cpu().numpy())
        y_prob.extend(probs.cpu().numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_prob = np.array(y_prob)

# -----------------------------------
# Confusion matrix
# -----------------------------------

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(7, 6))
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

ticks = np.arange(len(LABELS))
plt.xticks(ticks, LABELS, rotation=45)
plt.yticks(ticks, LABELS)

plt.xlabel("Predicted")
plt.ylabel("True")

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.tight_layout()
plt.show()

print("Confusion Matrix:\n", cm)

# -----------------------------------
# Metrics
# -----------------------------------

acc = accuracy_score(y_true, y_pred)

precision = precision_score(
    y_true, y_pred, average="macro", zero_division=0
)

recall = recall_score(
    y_true, y_pred, average="macro", zero_division=0
)

f1 = f1_score(
    y_true, y_pred, average="macro", zero_division=0
)

# ROC-AUC (multiclass, one-vs-rest)
n_classes = len(LABELS)
y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))

roc_auc = roc_auc_score(
    y_true_bin,
    y_prob,
    average="macro",
    multi_class="ovr"
)

# -----------------------------------
# Result table
# -----------------------------------

results_df = pd.DataFrame({
    "Metric": [
        "Accuracy",
        "Precision (macro)",
        "Recall (macro)",
        "F1-score (macro)",
        "ROC-AUC (macro, OVR)"
    ],
    "Value": [
        acc,
        precision,
        recall,
        f1,
        roc_auc
    ]
})

results_df
```

```python
import numpy as np
import librosa
import torch
from transformers import AutoProcessor, AutoModelForAudioClassification

MODEL_NAME = "MIT/ast-finetuned-audioset-10-10-0.4593"

SR = 16000
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

processor = AutoProcessor.from_pretrained(MODEL_NAME)
model = AutoModelForAudioClassification.from_pretrained(MODEL_NAME).to(DEVICE)
model.eval()

print("Model loaded on", DEVICE)
```

```python
THREAT_MAP = {

    "gunshot": [
        "Gunshot, gun fire",
        "Machine gun",
        "Rifle, shotgun"
    ],

    "siren": [
        "Siren",
        "Civil defense siren",
        "Ambulance siren",
        "Police siren",
        "Fire engine siren"
    ],

    "glass_break": [
        "Breaking glass"
    ],

    "distress_call": [
        "Screaming",
        "Shout",
        "Yell",
        "Crying, sobbing"
    ]
}
```

```python
id2label = model.config.id2label
```

```python
def detect_realtime_audio(
    audio_path,
    win_sec=2.0,
    hop_sec=1.0
):

    y, _ = librosa.load(audio_path, sr=SR, mono=True)

    win = int(win_sec * SR)
    hop = int(hop_sec * SR)

    segments = []

    for s in range(0, max(len(y) - win + 1, 1), hop):

        seg = y[s:s + win]

        if len(seg) < win:
            seg = np.pad(seg, (0, win - len(seg)))

        segments.append(seg.astype(np.float32))

    inputs = processor(
        segments,
        sampling_rate=SR,
        return_tensors="pt",
        padding=True
    )

    x = inputs["input_values"].to(DEVICE)

    with torch.no_grad():
        outputs = model(x)
        probs = torch.sigmoid(outputs.logits)

    # aggregate over time windows
    final_probs, _ = torch.max(probs, dim=0)

    result = {}

    for threat, names in THREAT_MAP.items():

        scores = []

        for i, label_name in id2label.items():
            if label_name in names:
                scores.append(final_probs[i].item())

        result[threat] = max(scores) if len(scores) else 0.0

    return result
```
